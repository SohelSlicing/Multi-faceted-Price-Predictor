# Model training for car prices

## Importing libraries and data

```python
import pandas as pd
import numpy as np
```

```python
def addcol(name, size, dataset):
    brand = [name] * size[0]
    dataset['Brand'] = brand
    dataset.head()
```

```python
#The data is present in different cvs files for different brands i'll combine all the data and add a column for brandname
audi = pd.read_csv('Data/Car price dataset/audi.csv')
addcol('Audi', audi.shape, audi)
audi.head()
```

```python
bmw = pd.read_csv('Data/Car price dataset/bmw.csv')
addcol('BMW', bmw.shape, bmw)
bmw.head()
```

```python
mercedes = pd.read_csv('Data/Car price dataset/cclass.csv')
addcol('Mercedes', mercedes.shape, mercedes)
mercedes.head()
```

```python
focus = pd.read_csv("Data/Car price dataset/focus.csv")
addcol('focus', focus.shape, focus)
focus.head()
```

```python
ford = pd.read_csv('Data/Car price dataset/ford.csv')
addcol('Ford', ford.shape, ford)
ford.head()
```

```python
hyundai = pd.read_csv('Data/Car price dataset/hyundi.csv')
addcol('Hyundai', hyundai.shape, hyundai)
hyundai.head()
```

```python
mercedes = pd.read_csv('Data/Car price dataset/merc.csv')
addcol("Mercedes", mercedes.shape, mercedes)
mercedes.head()
```

```python
skoda = pd.read_csv('Data/Car price dataset/skoda.csv')
addcol('Skoda', skoda.shape, skoda)
skoda.head()
```

```python
toyota = pd.read_csv('Data/Car price dataset/toyota.csv')
addcol('Toyota', toyota.shape, toyota)
toyota.head()
```

```python
# we have all the car in seprate dataframes lets combine them into 1 dataset

combined_dataset = [audi, bmw, mercedes, focus, ford, hyundai, skoda, toyota]
dataset = pd.concat(combined_dataset)
dataset.head()
```

```python
dataset.info()
```

```python
dataset = dataset.drop(columns='tax')
dataset = dataset.drop(columns='tax(£)')
```

```python
# milage is a important aspect of choosing for a car
dataset = dataset.dropna()
dataset = dataset.reset_index()
```

```python
dataset = dataset.drop(columns='index')
```
